```python
#Onboarding Day 4 (Pandas)
import pandas as pd
import numpy as np

#path = '/Users/paulinareina/Desktop/DataFiles/'
energydata = pd.read_csv('FY2020.csv')

#Question 1
companies = [x for x in energydata['Company Name']]
print('Question 1: How many different companies are represented in the data set?')
print('Answer 1:')
print(len(companies))

#Question 2
equeens = energydata[energydata['Borough']== 'Queens']
equeens_jobs = equeens['Job created'].dropna()
total = equeens_jobs.sum()
print('Question 2: What is the total number of jobs created for businesses in Queens?')
print('Answer 2:')
print(total)
energydata

#Question 3
emails = [x for x in energydata['company email']]

#cleaning nan from list using pandas isnull
emails_clean = [email for email in emails if not pd.isnull(email)]
#print(emails_clean)

emails_clean2 = []
domains = []
for email in emails_clean:
    if '@' in email:
        emails_clean2.append(email)
    else:
        domains.append(email)

inx = []
for email in emails_clean2:
    for i in range(len(email)):
        if email[i] == '@':
            inx.append(i)

for i in range(len(emails_clean2)):
    email = emails_clean2[i]
    upper = inx[i]+1
    domain = email[upper:]
    domains.append(domain)

domains_df = pd.DataFrame(domains)
#domains_df

#for col in domains_df.columns:
    #print(col)
    #print(type(col))
    
unique_tot = domains_df[0].unique()
print('Question 3: How many different unique email domains names are there in the data set?')
print('Answer 3:')
print(len(unique_tot))
    
#Question 4
#unique_nta = energydata['NTA'].unique()
#print(unique_nta)

ntas = [x for x in energydata['NTA']]
#print(nta)

nta_list = []
cnt_list = []
for nta in ntas:
    if nta not in nta_list:
        nta_list.append(nta)

for nta in nta_list:
    cnt = 0
    for nta2 in ntas:
        if nta == nta2:
            cnt = cnt + 1
    cnt_list.append(cnt)

nta_5 = []
cnt_new = []
for i in range(len(nta_list)):
    if cnt_list[i] >= 5:
        nta_5.append(nta_list[i])
        cnt_new.append(cnt_list[i])

#print(cnt_new)
#print(nta_5)

#filter data frame based on nta_5
nta_df = energydata[energydata['NTA'].isin(nta_5)]
nta_avg_sav = nta_df.groupby(['NTA'])['Total Savings'].mean()
nta_tot_jobs = nta_df.groupby(['NTA'])['Job created'].sum()
print('Question 4: Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?')
print('Answer 4:')
print('Average Total Savings')
print(nta_avg_sav)
#print('----------')
print('Total Number of Jobs')
print(nta_tot_jobs)

nta_final = pd.DataFrame(nta_avg_sav)
nta_final = nta_final.rename(columns={'Total Savings':'Avg Total Savings'})
nta_final['Total Jobs Created'] = nta_tot_jobs

nta_final.to_csv('Onboarding Day4.csv')

#Onboarding Day 5 (Plotting)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt

data = pd.read_csv('Onboarding Day4.csv')

energydata = pd.read_csv('FY2020.csv')

jobs = [x for x in data['Total Jobs Created']]
savings = [x for x in data['Avg Total Savings']]

#print('Scatter plot of jobs created versus average savings. Use both a standard and a logarithmic scale for the average savings')
plt.plot(savings,jobs, '.', color='darkviolet', markersize='10');
plt.xlabel('Average savings');
plt.ylabel('Jobs created');
plt.xlim(0, 500000);
plt.ylim(-1, 21);
plt.title('Jobs created vs. Average savings');
plt.show()

plt.plot(savings, jobs, '.', color='blue', markersize='10');
plt.xlabel('Average savings (log scale)');
plt.ylabel('Jobs created');
plt.title('Jobs created vs. Average savings (log scale)');
plt.xscale('log')
#plt.xlim(0, 1000000);
plt.show()

#source: https://stackoverflow.com/questions/47850202/plotting-a-histogram-on-a-log-scale-with-matplotlib
def plot_loghist(x, bins, c):
    hist, bins = np.histogram(x, bins=bins)
    logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
    plt.hist(x, bins=logbins, color=c)
    plt.xscale('log')

#print('Histogram of the log of the average total savings')
plt.figure(1)
plot_loghist(savings, 10, 'green')
plt.title('Histogram of the log of the average total savings');

#print('Line plot of the total jobs created for each month')
total_jobs = [x for x in energydata['Job created']]
date = [x for x in energydata['Effective Date']]

new_df = pd.DataFrame(total_jobs)
new_df = new_df.rename(columns={0:'Jobs'})
new_df['Date'] = date
new_df2 = new_df.dropna().reset_index()
new_df2['Date'] = pd.to_datetime(new_df2['Date'])
new_df2['Month'] = new_df2['Date'].dt.month
ser = new_df2.groupby(['Month'])['Jobs'].sum()
new_df3 = pd.DataFrame(ser).reset_index()
new_df3['Month Name'] = pd.to_datetime(new_df3['Month'], format='%m').dt.month_name().str[:3]

plt.figure(2)
plt.plot(new_df3['Month Name'], new_df3['Jobs'], color='orangered')
plt.title('Line plot of total jobs created for each month')
plt.xlabel('Month')
plt.ylabel('Total jobs');




```
